####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
tab_cross = pd.crosstab(df['GESTACAO'], df['APGAR5'])
tab_cross
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
tab_cross_pctlin = tab_cross.div(tab_cross.sum(axis = 1), axis = 0)
tab_cross_pctlin
```

```python
df_filtrado = df[(np.isfinite(df['PESO'])) & (np.isfinite(df['APGAR5']))]
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_filtrado, x='PESO', y='APGAR5')
plt.title('Relação entre Peso do Bebê e APGAR5')
plt.xlabel('Peso do Bebê (g)')
plt.ylabel('APGAR5')
plt.show()
```

Observando o gráfico de dispersão, podemos ver que existe uma leve tendência positiva entre o peso do bebê e o índice APGAR5. Isso significa que, em geral, bebês com maior peso tendem a ter valores de APGAR5 um pouco maiores, indicando melhores condições de saúde ao nascer.

O peso do bebê pode ser um fator que contribui para um APGAR5 mais alto, mas não é o único fator determinante. Outros fatores, como a saúde da mãe durante a gestação, complicações no parto e cuidados neonatais, também podem influenciar o índice APGAR5.

Sendo assim, podemos concluir que existe uma relação entre o peso do bebê e o APGAR5, mas essa relação é fraca e não explica completamente as variações no índice APGAR5.
